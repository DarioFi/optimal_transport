```python
import matplotlib.pyplot as plt
import numpy as np

# Set up the figure and axis
fig, ax = plt.subplots()

# Define the unit circle
# circle = plt.Circle((0, 0), 1, color='black', fill=False, linestyle='--')

# Add the circle to the plot
# ax.add_artist(circle)

start_ang = np.deg2rad(30)
# Set the angle theta and beta
theta = np.deg2rad(70)  # 30 degrees
beta = np.deg2rad(45)  # 45 degrees

# Define points on the circle
p_1 = np.array([np.cos(start_ang), np.sin(start_ang)])
p_2 = np.array([np.cos(start_ang + theta), np.sin(start_ang + theta)])
v = np.array([np.cos(start_ang + beta), np.sin(start_ang + beta)])

# Plot the vectors
ax.quiver(*p_1, -p_1[0], -p_1[1], angles='xy', scale_units='xy', scale=1, color='blue', label=r'$A^+$')
ax.quiver(0,0, p_2[0], p_2[1], angles='xy', scale_units='xy', scale=1, color='blue', label=r'$A^-$')
ax.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='blue', label=r'$v$')

# Annotate the vectors
ax.text(p_1[0] * 1.1, p_1[1] * 1.1, r'$p_1$', fontsize=12, ha='center')
ax.text(p_2[0] * 1.1, p_2[1] * 1.1, r'$p_2$', fontsize=12, ha='center')
ax.text(v[0] * 1.1, v[1] * 1.1, r'$v$', fontsize=12, ha='center')

# add text next to vectors with m_1 and m_2
off = 0.1
ax.text(p_1[0] * .6 + off, p_1[1] * .5 - 1.4 * off, r'$m_1$', fontsize=12, ha='center')
ax.text(p_2[0] * .5 - 1.8 * off, p_2[1] * .5, r'$m_2$', fontsize=12, ha='center')

ax.text(0, -.1, r'$s$', fontsize=12, ha='center', va='center')

# Draw the angles theta and beta
from matplotlib.patches import Arc

# Parameters for the arcs
arc_radius = 0.3  # Radius of the arcs
arc_angle_1 = np.rad2deg(beta)  # Angle between A+ and A-
arc_angle_2 = np.rad2deg(theta - beta)  # Angle between A- and v

arc1 = Arc([0, 0], 2 * arc_radius, 2 * arc_radius, angle=0,
           theta1=np.rad2deg(start_ang), theta2=np.rad2deg(start_ang + theta), color='black')
ax.add_patch(arc1)

arc2 = Arc([0, 0], 3 * arc_radius, 3 * arc_radius, angle=0,
           theta1=np.rad2deg(start_ang), theta2=np.rad2deg(start_ang + beta), color='black')
ax.add_patch(arc2)

# Annotate the angles
ax.text(2 * arc_radius * np.cos(start_ang + beta / 2), 2 * arc_radius * np.sin(start_ang + beta / 2),
        r'$\beta$', fontsize=12,
        ha='center', va='center')

ax.text(1.3 * arc_radius * np.cos(start_ang + beta + (theta - beta) / 2),
        1.3 * arc_radius * np.sin(start_ang + beta + (theta - beta) / 2),
        r'$\theta$', fontsize=12,
        ha='center', va='center')

# Set the aspect ratio of the plot to be equal
ax.set_aspect('equal')

# Set the limits of the plot
ax.set_xlim(-1.5, 1.5)
ax.set_ylim(-1.5, 1.5)

# Set grid and labels
ax.set_xlabel('x')
ax.set_ylabel('y')

# hide axis and rectangle around the plot
ax.axis('off')
ax.set_frame_on(False)

# Display the plot
plt.show()
```

```python
# Set up the figure and axis
fig, ax = plt.subplots()

# Define the angle theta (for example 70 degrees)
theta = np.deg2rad(70)

start_ang_deg = 30
start_ang = np.deg2rad(start_ang_deg)
# Define points for the two vectors
p1 = np.array([np.cos(start_ang), np.sin(start_ang)])  # Start at (1, 0)
p2 = np.array([np.cos(theta), np.sin(theta)])  # Point at an angle theta

# Plot the vectors from the origin to p1 and p2
ax.quiver(0, 0, p1[0], p1[1], angles='xy', scale_units='xy', scale=1, color='blue', label=r'$p_1$')
ax.quiver(0, 0, p2[0], p2[1], angles='xy', scale_units='xy', scale=1, color='blue', label=r'$p_2$')

# Annotate the vectors
ax.text(p1[0] * 1.1, p1[1] * 1.1, r'$m_1$', fontsize=12, ha='center')
ax.text(p2[0] * 1.1, p2[1] * 1.1, r'$m_2$', fontsize=12, ha='center')

# Draw the angle arc between p1 and p2
arc_radius = 0.3  # Radius of the angle arc
arc = Arc([0, 0], 2 * arc_radius, 2 * arc_radius, angle=start_ang_deg,
          theta1=0, theta2=np.rad2deg(theta) - start_ang_deg, color='black')
ax.add_patch(arc)

# Annotate the angle
ax.text(1.5 * arc_radius * np.cos((theta + start_ang) / 2), 1.5 * arc_radius * np.sin((theta + start_ang) / 2),
        r'$\theta$', fontsize=12, ha='center', va='center')

# Set the aspect ratio of the plot to be equal
ax.set_aspect('equal')

# Set the limits of the plot
ax.set_xlim(-1.5, 1.5)
ax.set_ylim(-1.5, 1.5)

# remove axis and rectangle around the plot
ax.axis('off')
ax.set_frame_on(False)

# Display the plot
plt.show()
```
