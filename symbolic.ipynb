```python
import sympy as sp
```

```python
# Define the symbols
x, y = sp.symbols('x y')
beta = sp.symbols('beta')

# Compute theta_L and theta_R based on beta
theta_L = -beta / 2
theta_R = beta / 2

# Coordinates of L and R on the unit circle
L = (sp.cos(theta_L), sp.sin(theta_L))
R = (sp.cos(theta_R), sp.sin(theta_R))

# Euclidean distances
d1 = sp.sqrt(x ** 2 + y ** 2)
d2 = sp.sqrt((x - L[0]) ** 2 + (y - L[1]) ** 2)
d3 = sp.sqrt((x - R[0]) ** 2 + (y - R[1]) ** 2)

# Total length function S
S = d1 + d2 + d3

# Derivatives of S with respect to x and y
dS_dx = sp.diff(S, x)
dS_dy = sp.diff(S, y)

# Display the results
S, dS_dx, dS_dy
```

```python
# compute dS_dx and dS_dy at the point (x, y) = (0, 0)
dS_dx.subs({x: 0, y: 0}), dS_dy.subs({x: 0, y: 0})
```
